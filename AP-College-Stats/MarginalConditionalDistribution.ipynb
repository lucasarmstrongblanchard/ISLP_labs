* Marginal distribution: focuses on one dimension in a joint distribution
* Take the below dataframe as an example--the column represents time spent studying, and the index represents the score on a test

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

time_studied = np.array([[0.2, 0, 4, 16, 20], [0.4, 0, 20, 30, 10], [0.6, 2, 4, 32, 32]])
cols = ["score", "0-20", "21-40", "41-60", "60-100"]
df = pd.DataFrame(time_studied, columns=cols)
df.set_index("score", inplace=True)


df.style
```

```python
def pretty_print_table(df, title, row_labels=None, col_labels=None):
    fig = plt.figure(figsize = (8, 2))
    ax = fig.add_subplot(111)

    ax.table(cellText = df.values,
          rowLabels = row_labels,
          colLabels = col_labels,
          loc = "center"
         )
    ax.set_title(title)

    ax.axis("off");
```

```python
# To get the marginal distribution of the percent correct--answers the question what percentage 
# of students got x score
total_correct = np.array(df.sum(axis=1))
marginal_dis = total_correct/sum(total_correct)
marginal_total = pd.DataFrame([marginal_dis], columns=df.index.values)
title="marginal distribution of score"
pretty_print_table(marginal_total, title, col_labels=marginal_total.columns)

# marginal distribution for time studied 
time_studied = np.array(df.sum(axis=0))
marginal_time_studied = pd.DataFrame([time_studied/sum(time_studied)], columns=cols[1:])
title="marginal distribution of time studied"
pretty_print_table(marginal_time_studied, title, col_labels=marginal_time_studied.columns)
```

* Conditioanl distribution: distribution of percent correct given that students studied between 41 and 60 minutes

```python
studied = np.array(df["41-60"])
conditional_dis = pd.DataFrame([studied / sum(studied)], columns=df.index.values)
title = "Conditional Distribution  of Percent Correct Given a Study Time of 41 to 60 Minutes"
pretty_print_table(conditional_dis, title, col_labels=conditional_dis.columns)
```
